```python
import netCDF4

import numpy as np
import matplotlib.pyplot as plt
```

```python
dir_data = '../data/'
file_tg = dir_data + 'tg_ens_mean_0.25deg_reg_v30.0e.nc'
file_pp = dir_data + 'pp_ens_mean_0.25deg_reg_v30.0e.nc'
file_tn = dir_data + 'tn_ens_mean_0.25deg_reg_v30.0e.nc'
file_tx = dir_data + 'tx_ens_mean_0.25deg_reg_v30.0e.nc'
file_qq = dir_data + 'qq_ens_mean_0.25deg_reg_v30.0e.nc'
```

```python
nc_tg = netCDF4.Dataset(file_tg,'r')
nc_pp = netCDF4.Dataset(file_pp,'r')
nc_tn = netCDF4.Dataset(file_tn,'r')
nc_tx = netCDF4.Dataset(file_tx,'r')
nc_qq = netCDF4.Dataset(file_qq,'r')
```

```python
nc_pp.variables.keys()
```

```python
nc_pp.variables['time'][42]
```

```python
# get the latitute and longitude grid
lat = nc_qq.variables['latitude'][:]
lon = nc_qq.variables['longitude'][:]
```

```python
lat.shape, lon.shape
```

```python
idx = -10
pp = nc_pp.variables['pp'][idx]
date = nc_pp.variables['time'][idx]

qq = nc_qq.variables['qq'][idx]
tg = nc_tg.variables['tg'][idx]
tn = nc_tn.variables['tn'][idx]
tx = nc_tx.variables['tx'][idx]
```

```python
grid = np.meshgrid(lon,lat)

plt.figure(figsize=(5,5))
plt.title('Pressure ' + str(date))
plt.pcolormesh(grid[0],grid[1],pp)
plt.colorbar()
plt.xlim(-11, 28)
plt.ylim(35, 72)
plt.show()

plt.figure(figsize=(5,5))
plt.title('Radiation ' + str(date))
plt.pcolormesh(grid[0],grid[1],qq)
plt.colorbar()
plt.xlim(-11, 28)
plt.ylim(35, 72)
plt.show()

plt.figure(figsize=(5,5))
plt.title('Temperature ' + str(date))
plt.pcolormesh(grid[0],grid[1],tg)
plt.colorbar()
plt.xlim(-11, 28)
plt.ylim(35, 72)
plt.show()

plt.figure(figsize=(5,5))
plt.title('Min Temperature ' + str(date))
plt.pcolormesh(grid[0],grid[1],tn)
plt.colorbar()
plt.xlim(-11, 28)
plt.ylim(35, 72)
plt.show()

plt.figure(figsize=(5,5))
plt.title('Max Temperature ' + str(date))
plt.pcolormesh(grid[0],grid[1],tx)
plt.colorbar()
plt.xlim(-11, 28)
plt.ylim(35, 72)
plt.show()
```

```python
# get the average temperature over the time axis
tg_mean = np.mean(tg, axis=0)
```

```python
grid = np.meshgrid(lon,lat)

plt.contourf(grid[0],grid[1],tg_mean)
plt.colorbar()
```
